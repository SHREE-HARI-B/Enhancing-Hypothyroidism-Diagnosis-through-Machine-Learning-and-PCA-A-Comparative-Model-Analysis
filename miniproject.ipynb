# Preprocessing the dataset - doing complete preprocessing until the dataset becomes completely clean and usable

```python
df.sex.replace({'F':2,'M':1},inplace=True)
round_Values = round(df.sex.mean())
df.sex.fillna(round_Values,inplace=True)
df.sex.unique()
```

# Checking corelation because PCA works using corelation concept

```python
df.corr()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
cols = df.select_dtypes(include=['object'])
for i in cols.columns:
    try:
        df[i] = le.fit_transform(df[i])
    except:
        continue
```

# Analysing which column to be eliminated to give correct correlation values between 0 and 1

```python
for a in range(len(df.corr())):
    for b in range(a):
        if((df.corr().iloc[a,b]) >= 0.7):
            print(df.corr().columns[b])
```

# Dropping the above mentioned column

```python
df.drop('TT4',axis=1,inplace=True)
```

# Plotting Histograms to find the range of correlation

```python
df.hist(bins=25,figsize=(20,20));
```

# Since correlation values are more than 1 , we use PCA to get correlation values between 0 and 1

```python
X = df.drop('target',axis=1)
y = df.target
df2 = X 
```

```python
y.unique()
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=10)
```

```python
v = pca.fit_transform(X)
```

```python
X_pca = pd.DataFrame(data = v, columns = ['component_1', 'component_2', 'component_3', 'component_4', 'component_5', 'component_6', 'component_7', 'component_8', 'component_9', 'component_10'])
```

```python
X_pca
```

# Now plotting histogram having values between 0 and 1 also X_pca values with negative values gets resolved 

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python
for i in X_pca.columns:
    X_pca[i] = scaler.fit_transform(X_pca[[i]])
```

```python
X_pca.hist(bins=25,figsize=(20,20));
```

```python
X_pca
```

# Splitting the data into training data and testing data to predict the accuracy

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X_pca,y,test_size=0.33,random_state=42)
```

```python
from sklearn.metrics import accuracy_score
```

# Predicting accuracy using Decision Tree algorithm without PCA

```python
from sklearn.tree import DecisionTreeClassifier
tree = DecisionTreeClassifier(max_depth=3)
clf = tree.fit(X_train,y_train)
treepredict = clf.predict(X_test)
accuracy_score(treepredict,y_test)
```

# Predicting accuracy using Random Forest algorithm without PCA

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(max_depth=2,n_estimators=200)
rclf = rf.fit(X_train,y_train)
rfpred = rclf.predict(X_test)
accuracy_score(rfpred,y_test)
```

# Predicting accuracy using KNN algorithm without PCA

```python
from sklearn.neighbors import KNeighborsClassifier
neigh = KNeighborsClassifier(n_neighbors=3)
knnclf = neigh.fit(X_train,y_train)
y_pred = knnclf.predict(X_test)
accuracy_score(y_pred,y_test)
```

# Predicting accuracy using SVC algorithm without PCA

```python
from sklearn.svm import SVC
svm = SVC(kernel="sigmoid")
sclf = svm.fit(X_train,y_train)
y_pred = sclf.predict(X_test)
accuracy_score(y_pred,y_test)
```

# Predicting accuracy using Logistic Regression algorithm without PCA

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(max_iter=1000)
lrclf = lr.fit(X_train,y_train)
y_pred = lrclf.predict(X_test)
accuracy_score(y_pred,y_test)
```
